```python
from dotenv import load_dotenv
load_dotenv()

from graph.state import GraphState
state = GraphState(question="Who does O'Brien represent as a historical figure? In the context of the 20th century? what about the 21st century?", generation="L6", web_search=True, documents=[])
```

```python
from graph.nodes.retrieve import retrieve
retrieve_result = retrieve(state)
print(retrieve_result)
print('Results', retrieve_result)
print(retrieve_result['documents'])

state.update(retrieve_result)
print('State keys: ', state.keys())
print(state)
```

```python
from graph.nodes.grade_documents import grade_documents
filtered_docs = grade_documents(state)
print('Results', filtered_docs)
print(filtered_docs['documents'])

state.update(filtered_docs)
print('State keys: ', state.keys())
print(state)
```

```python
from graph.nodes.web_search import web_search
web_search_result = web_search(state)
print('Results', web_search_result)
print(web_search_result['documents'][-1].page_content)

state.update(web_search_result)
print('State keys: ', state.keys())
print(state)
```

```python
from graph.nodes.generate import generate
generate_result = generate(state)
print('Results', generate_result)
print(generate_result['generation'])

state.update(generate_result)
print('State keys: ', state.keys())
print(state)
```
